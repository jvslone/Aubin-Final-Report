## Interference Patterns

Below shows the level of the amplitude of the resulting wave when two microwave sine waves interfere, the corrected term sets the amplitude of each of the signals to be equal \
This shows how an imbalance in the two combining signal's amplitude (such as a power distribution imbalance between slots in a QH) adds error to the amplitude of the resulting combined signal \
This serves to quantify the relationship between relative amplitude noise and noise on the final measurement

```python
#Parameters
time_data_points = 1000
phase_data_points = 1000
times = np.linspace(-0.9999,0.9999,time_data_points)
phases = np.linspace(0,360,phase_data_points)
percent = 0.1 #Percent difference in signal amplitudes

f1 = 1
A1 = 1
S1 = lambda t: A1*np.sin(np.pi*f1*t)

f2 = 1
A2 = 1 + percent/100
S2 = lambda t, theta: A2*np.sin(np.pi*(f2*t + theta/180))
S2corr = lambda t, theta: A1*np.sin(np.pi*(f2*t + theta/180))

maxnon0 = []
maxcorr0 = []

#Calculation
for phase in phases:
  ydata1 = S1(times)
  ydata2 = S2(times, phase)
  ydata3 = S2corr(times, phase)

  totalnon = ydata1 + ydata2
  totalcorr = ydata1 + ydata3

  maxnon = totalnon.max()
  maxcorr = totalcorr.max()

  maxnon0.append(maxnon)
  maxcorr0.append(maxcorr)

maxnon0 = np.array(maxnon0)
maxcorr0 = np.array(maxcorr0)

differences = 100*(maxnon0-maxcorr0)/(maxnon0)
```

```python
#Plotting
fig7 = plt.figure(figsize=(10,6))
plt.plot(phases,maxnon0,'-r',label='Non-Corrected')
plt.plot(phases,maxcorr0,'-b',label='Corrected')

plt.xlabel('Phase Difference(deg)')
plt.ylabel('Combined Amplitude')
plt.title('Corrected vs Non-Corrected Signals')

plt.legend()
plt.show()
```

This plot shows the percent difference between the corrected and non-corrected signals

```python
#Plotting
fig8 = plt.figure(figsize=(10,6))

plt.plot(phases,differences,'-r',label='Diff')
mindiff = differences.min()
print(mindiff)
print(0.0001)

plt.xlabel('Phase Difference(deg)')
plt.ylabel('Difference(%)')
plt.title('Percent Diff')

plt.ylim(0,mindiff*2)

plt.legend()
plt.show()
```

Here I am using the data from the first plot (the simple detector curve) to try and translate the detector reading back to an input power by inverting the relation, using both the 'ideal fit' and the raw data

```python
from scipy.interpolate import interp1d
func646 = interp1d(det646curve,det646Power,kind='cubic')

func646ideal = lambda x: (np.log((x-2.846)/9.353))/0.1851

curve_corr0 = func646ideal(Det646_mult)
curve_corr1 = func646(Det646_mult)
mW = lambda dBm: 10**((dBm+1.15)/10)
curve_corr2 = mW(curve_corr1)
curve_corr0p = mW(curve_corr0)

fig19 = plt.figure(figsize=(20,8))

plt.plot(Phase_mult,curve_corr1,'-r',label='Corrected dBm')
plt.plot(Phase_mult,curve_corr0,'--k',label='Approximated dBm')


plt.xlabel('Phase(Deg)')
plt.ylabel('Power(dBm)')
plt.title("Multiple Periods Corrected Source Analysis")
plt.legend()
plt.grid()
plt.show()
```
